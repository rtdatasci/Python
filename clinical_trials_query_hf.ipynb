The primary objective of this Python script is to demonstrate a multi-step clinical trial metadata extraction pipeline using the smolagents library. Specifically, it aims to:

Search for a Clinical Trial: Use an initial query (e.g., "Ewing Sarcoma clinical trials completed phase 2") to find an official study page on ClinicalTrials.gov and extract its NCT ID and Official Title. (Example found: NCT01583543).

Preprocess Eligibility Criteria: Take a known clinical trial ID (NCT05275426) and use a Web Search Agent to find its eligibility criteria.

Format the Data: Convert the raw eligibility text into a unified JSON format for easier machine processing and analysis.

ğŸ› ï¸ Resources Used
The script relies on the following key resources and tools:

smolagents Library: This is the core framework used to create and run specialized agents (CodeAgent) for search and data processing tasks.

Hugging Face API Token (HF_TOKEN): The agents use models from the Hugging Face Inference Client (InferenceClientModel) for their reasoning and code generation, which requires an API key stored as a Colab secret.

WebSearchTool (Google Search): This tool, available to the agents, is used to query the web, primarily targeting the ClinicalTrials.gov website to locate study details and eligibility criteria.

ClinicalTrials.gov: This is the ultimate data source, as the agents are instructed to specifically search for and parse the official study pages for clinical trial metadata.

```python
from google.colab import userdata

# Ensure you have your Hugging Face API token stored in Colab secrets under the name 'HF_TOKEN'
HF_TOKEN = userdata.get('HF_TOKEN')
```

```python
from smolagents import CodeAgent, InferenceClientModel, WebSearchTool

# Tools available to all agents
TOOLS = [WebSearchTool()]

# --- 1. Clinical Trial Search Agent ---
search_agent = CodeAgent(
    tools=TOOLS,
    model=InferenceClientModel(api_key=HF_TOKEN),
    instructions="""You are a clinical trial search agent.
    Your task is to formulate a precise web search query to find the
    official 'ClinicalTrials.gov' study page for the given condition.
    The output should be the study's **NCT ID** (e.g., NCT05275426) and its **Official Title**."""
)

# # --- Define the initial query ---
# ingestion_query = "Ewing Sarcoma clinical trials completed phase 2"

# # --- Run the first step (example) ---
# # Note: You would run this and parse the output for the NCT ID.
# # For this example, we will manually use the ID found on the open page.
# nct_id = "NCT05275426"
# trial_title = "A Phase II Study of Oral CHK1 Inhibitor LY2880070 in Combination With Low-Dose Gemcitabine in Patients With Relapsed or Refractory Ewing Sarcoma, Ewing-like Sarcoma, and Desmoplastic Small Round Cell Tumor"
```

```python
# --- Define the initial query ---
ingestion_query = "Ewing Sarcoma clinical trials completed phase 2"

ingestion_result = search_agent.run(ingestion_query)
print(ingestion_result)
```

```python
# --- 2. Eligibility Preprocessing Agent ---
nct_id = "NCT05275426"
trial_title = "A Phase II Study of Oral CHK1 Inhibitor LY2880070 in Combination With Low-Dose Gemcitabine in Patients With Relapsed or Refractory Ewing Sarcoma, Ewing-like Sarcoma, and Desmoplastic Small Round Cell Tumor"

preprocessing_agent = CodeAgent(
    tools=TOOLS,
    model=InferenceClientModel(api_key=HF_TOKEN),
    instructions=f"""You are a clinical trials metadata preprocessing agent.
    Your task is to take a ClinicalTrials.gov ID (e.g., {nct_id}) and use your
    WebSearchTool to find the associated study page, focusing on the
    **'Participation Criteria'** section.

    Your goal is to parse this raw text and convert it into a **unified JSON format**.
    You must extract the information from the 'Eligibility Criteria' section.

    Use the following specific JSON schema for the output:
    {{
      "clinical_trials_id": "{nct_id}",
      "title": "{trial_title}",
      "ages_eligible_for_study": ["Child", "Adult", "Older Adult", "..."],
      "performance_status": "Karnofsky/Lansky â‰¥70%",
      "disease_characteristics": "Histologically documented Ewing sarcoma as molecularly defined by...",
      "body_weight_criteria": "patients must be â‰¥40 kg...",
      "other_inclusion_criteria": [
        "Adequate bone marrow function: ANC â‰¥ 1500/mm3...",
        "Adequate renal function: eGFR rate â‰¥ 60 mL/min/1.73m2..."
      ]
    }}

    Only output the JSON object."""
)
```

```python
# --- Run Preprocessing Agent (Step 2 Example) ---
nct_id = "NCT05275426"
trial_title = "A Phase II Study of Oral CHK1 Inhibitor LY2880070 in Combination With Low-Dose Gemcitabine in Patients With Relapsed or Refractory Ewing Sarcoma, Ewing-like Sarcoma, and Desmoplastic Small Round Cell Tumor"

preprocessing_query = f"Parse the full eligibility criteria for ClinicalTrials.gov ID: {nct_id} and convert it to the unified JSON format."
print(f"--- Running Preprocessing Agent for {nct_id} ---")
preprocessing_result = preprocessing_agent.run(preprocessing_query)
print(preprocessing_result)
```
