```python
from smolagents import CodeAgent, InferenceClientModel, WebSearchTool

# Initialize the agent with the WebSearchTool from the Colab
agent = CodeAgent(
    tools=[WebSearchTool()],
    model=InferenceClientModel(),
    # Add instructions to guide the agent to formulate the best search query
    instructions="You are a real estate search assistant. Formulate a precise web search query to find listings on popular sites like Zillow or Realtor.com based on the user's criteria. Summarize the findings."
)

```

```python
# Run the agent with your natural language query
query = "single family house, 3 bedroom, near school zone with greatschoolratings above 8 in Austin, TX"
result = agent.run(query)
print(result)
# Result will be a summary of top search results, not structured house data.
```
