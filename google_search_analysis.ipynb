## Goal:
Check google search trends using pytreds (can pip install)

Used 'langchain' as query topic in this example. Since langchain is fairly new, all the searches appear after 2023 (see the final plot)



```python
#!pip install pytrends
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
# Can change the query topic in the payload: trends.build_payload(kw_list=["langchain"])
trends.build_payload(kw_list=["langchain"])
data = trends.interest_by_region()
data = data.sort_values(by="langchain", ascending=False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x="geoName",
                        y="langchain",
                        figsize=(15,12), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['langchain'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['langchain'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for langchain',
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
