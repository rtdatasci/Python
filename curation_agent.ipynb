1. Ingestion Agent (Initial Setup)
This first agent's job is to find the raw metadata for single-cell datasets from GEO. It will primarily use the WebSearchTool.

Objective: Formulate a precise search query to find GEO accession numbers or summaries for single-cell RNA-seq datasets.

```python
!pip install smolagents
```

```python
from smolagents import CodeAgent, InferenceClientModel, WebSearchTool

# Tools available to all agents
TOOLS = [WebSearchTool()]

# --- 1. Ingestion Agent ---
ingestion_agent = CodeAgent(
    tools=TOOLS,
    model=InferenceClientModel(),
    # Instructions to guide the agent for its specific task
    instructions="You are a data ingestion agent. Your task is to formulate a precise web search query to find summaries of 'single-cell RNA-seq' or 'scRNA-seq' datasets from the NCBI Gene Expression Omnibus (GEO) repository. The output should be a list of GEO accession numbers (e.g., GSEXXXXX) and their titles."
)

# --- Define the initial query ---
ingestion_query = "single-cell RNA-seq datasets GEO"

# --- Run the first step ---
print("--- Running Ingestion Agent ---")
ingestion_result = ingestion_agent.run(ingestion_query)
print(ingestion_result)
```

ğŸ› ï¸ 2. Preprocessing Agent: Metadata Normalization
Agent Definition
This agent needs clear instructions to search for a specific GEO accession number's detailed metadata page and then apply a standard structure to the extracted information.

```python
# Assuming you have the previous imports and TOOLS defined from the Ingestion step:
# from smolagents import CodeAgent, InferenceClientModel, WebSearchTool
# TOOLS = [WebSearchTool()]

# --- 2. Preprocessing Agent ---
preprocessing_agent = CodeAgent(
    tools=TOOLS,
    model=InferenceClientModel(),
    instructions="""You are a metadata preprocessing agent. Your task is to take a GEO accession number (e.g., GSEXXXXX) and use your WebSearchTool to find the associated GEO Summary page (GSE), which contains the raw metadata (e.g., study design, sample characteristics, overall design).

Your goal is to parse this raw text and convert it into a **unified JSON format**. Use the following specific JSON schema for the output:
{
  "accession_id": "GSEXXXXX",
  "title": "Study Title",
  "study_type": "single cell RNA-seq",
  "platform": "Illumina HiSeq, etc.",
  "organism": ["Homo sapiens", "Mus musculus"],
  "tissue_source": ["lung", "blood"],
  "cell_type": ["T cell", "macrophage"],
  "summary_description": "A concise summary of the dataset."
}
Only output the JSON object."""
)
```

```python
#Example run
# --- Simulate input from the Ingestion Agent (Step 1) ---
# In a real pipeline, ingestion_result would be parsed to get the ID.
# For now, we manually input a known GEO ID.
example_geo_id = "GSE139369"
preprocessing_query = f"Parse the full metadata for GEO accession: {example_geo_id} and convert it to the unified JSON format."

# --- Run Preprocessing Agent (Step 2) ---
print(f"--- Running Preprocessing Agent for {example_geo_id} ---")
preprocessing_result = preprocessing_agent.run(preprocessing_query)
print(preprocessing_result)
```

ğŸ”¬ 3. Ontology Mapping Agent: Term Standardization
Agent Definition
This agent needs to be instructed to specifically target biological terms within the JSON structure and then use its tools (in this case, still the WebSearchTool to simulate looking up ontology terms) to find the best match and assign a confidence score.

```python
# Assuming you have the previous imports and TOOLS defined.
# from smolagents import CodeAgent, InferenceClientModel, WebSearchTool
# TOOLS = [WebSearchTool()]

# --- 3. Ontology Mapping Agent ---
ontology_agent = CodeAgent(
    tools=TOOLS,
    model=InferenceClientModel(),
    instructions="""You are an ontology mapping agent. Your task is to receive a unified JSON metadata object.

    You must focus on the values for the **'tissue_source'** and **'cell_type'** fields.

    For each value (e.g., 'lung' or 'T cell'), use your WebSearchTool to find the corresponding term identifier (ID) and label from a relevant controlled vocabulary:
    * **Tissue/Organ:** Use UBERON (e.g., UBERON:0002103 for lung).
    * **Cell Type:** Use Cell Ontology (CL) (e.g., CL:0000084 for T cell).

    You must create a new JSON object that includes the original accession_id and replaces the original list values with an array of objects. Each new object must contain the following keys: **'original_term'**, **'mapped_ontology'**, **'mapped_ID'**, **'mapped_label'**, and **'confidence_score'** (a float between 0.0 and 1.0, where 1.0 is a perfect match).

    Only output the new JSON object.
    """
)
```

```python
#example run
# --- Simulate input from the Preprocessing Agent (Step 2) ---
# This is a sample unified JSON object (GSE139369)

### TEST ###
# unified_json_input = """
# {
#   "accession_id": "GSE139369",
#   "title": "Single-cell RNA-seq of lung tissue and bronchoalveolar lavage fluid from COVID-19 patients",
#   "study_type": "single cell RNA-seq",
#   "platform": "Illumina NextSeq 500",
#   "organism": ["Homo sapiens"],
#   "tissue_source": ["lung parenchyma", "bronchoalveolar lavage fluid"],
#   "cell_type": ["T-cells", "alveolar macrophages", "natural killer cells"],
#   "summary_description": "Analysis of immune response in severe COVID-19."
# }
# """

# ontology_mapping_query = f"Map the ontology terms for 'tissue_source' and 'cell_type' in this unified JSON metadata: {unified_json_input}"
### END OF TEST ###

ontology_mapping_query = f"Map the ontology terms for 'tissue_source' and 'cell_type' in this unified JSON metadata: { preprocessing_result }"
# --- Run Ontology Mapping Agent (Step 3) ---
print(f"--- Running Ontology Mapping Agent for GSE139369 ---")
ontology_result = ontology_agent.run(ontology_mapping_query)
print(ontology_result)
```

ğŸ§‘â€ğŸ’» 4. Review Coordination Agent: Human-in-the-Loop Prep
Agent Definition
This agent will take the complex, mapped JSON structure and re-format it into a highly scannable, Markdown-based review report. This leverages the agent's ability to summarize and re-structure information effectively.

```python
from smolagents import CodeAgent, InferenceClientModel

# --- 4. Review Coordination Agent ---
review_agent = CodeAgent(
    # Note: This agent doesn't need external tools; it only processes the previous agent's output.
    tools=[],
    model=InferenceClientModel(),
    instructions="""You are a review coordination agent. Your task is to receive the JSON output from the Ontology Mapping Agent.

    You must convert this JSON into a structured, human-readable review report using Markdown formatting.

    **Prioritize highlighting terms that require human review.** Specifically, create a section for terms where the 'confidence_score' is less than 0.95.

    The final output should be a single Markdown report with the following structure:

    ### Review Report: [accession_id]

    #### Overall Summary
    - Total Terms Mapped: [Count]
    - Terms Requiring Review (Confidence < 0.95): [Count]

    #### Full Mapping Details
    [List all original terms and their final mapped ontology ID/label and score.]

    #### âš ï¸ Terms Flagged for Review (Confidence < 0.95)
    [List only the terms where the confidence score is low, presenting the original term, mapped ID, and score clearly.]

    Only output the Markdown report."""
)
```

```python
#Example run
# --- Simulate input from the Ontology Mapping Agent (Step 3) ---
# Note: This is a simulated output showing the required structure and varied confidence scores.
mapped_json_input = """
{
  "accession_id": "GSE139369",
  "mapped_tissue_source": [
    {
      "original_term": "lung parenchyma",
      "mapped_ontology": "UBERON",
      "mapped_ID": "UBERON:0002048",
      "mapped_label": "pulmonary parenchyma",
      "confidence_score": 0.85
    },
    {
      "original_term": "bronchoalveolar lavage fluid",
      "mapped_ontology": "UBERON",
      "mapped_ID": "UBERON:0001878",
      "mapped_label": "respiratory tract fluid",
      "confidence_score": 0.98
    }
  ],
  "mapped_cell_type": [
    {
      "original_term": "T-cells",
      "mapped_ontology": "CL",
      "mapped_ID": "CL:0000084",
      "mapped_label": "T cell",
      "confidence_score": 1.0
    },
    {
      "original_term": "alveolar macrophages",
      "mapped_ontology": "CL",
      "mapped_ID": "CL:0000063",
      "mapped_label": "alveolar macrophage",
      "confidence_score": 0.96
    },
    {
      "original_term": "natural killer cells",
      "mapped_ontology": "CL",
      "mapped_ID": "CL:0000623",
      "mapped_label": "natural killer cell",
      "confidence_score": 1.0
    }
  ]
}
"""

review_query = f"Generate a human-readable review report based on the following mapped metadata JSON, highlighting terms with confidence less than 0.95: {mapped_json_input}"

# --- Run Review Coordination Agent (Step 4) ---
print("--- Running Review Coordination Agent ---")
review_report = review_agent.run(review_query)
print(review_report)
```
